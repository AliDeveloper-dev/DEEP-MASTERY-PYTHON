### Why does it exist?
Bohot saare problems mein brute force approach nested loop use karta hai — O(n²). Two Pointer ye same problems O(n) mein solve karta hai — sirf do pointers ko smartly move karke.

```
arr = [1, 2, 3, 4, 5]
       ↑               ↑
      left            right

left  → 0 se start (left side)
right → end se start (right side)
dono move karte hain conditions ke hisaab se
jab tak left < right
```

````python
arr = [1, 2, 3, 4, 5, 6, 7, 8, 9]
target = 12

for i in range(len(arr)):
    for j in range(i+1, len(arr)):
        if arr[i] + arr[j] == target:
            print(f"Pair found: {arr[i]}, {arr[j]}")
````

# Two Pointer O(n):

````python
arr = [1, 3, 5, 7, 9]   # SORTED hona zaroori hai
target = 12

left = 0
right = len(arr) - 1

while left < right:
    current_sum = arr[left] + arr[right]
    
    if current_sum == target:
        print(f"Pair found: {arr[left]}, {arr[right]}")
        break
    elif current_sum < target:
        left += 1     # sum chota hai — left aage badho
    else:
        right -= 1    # sum bada hai — right peeche aao
````

````python
def is_palindrome(s):
    left = 0
    right = len(s) - 1
    
    while left < right:
        if s[left] != s[right]:
            return False    # match nahi — palindrome nahi
        left += 1
        right -= 1
    
    return True    # sab match — palindrome hai

print(is_palindrome("racecar"))   # True
print(is_palindrome("hello"))     # False
print(is_palindrome("madam"))     # True
````

````python
def is_palindrome(s):
    left = 0
    right = len(s)-1
    while left < right:
        if s[left] != s[right]:
            return False
        left += 1
        right -=1
    return True
print(is_palindrome("level"))
````

````python
arr = [2, 1, 5, 1, 3, 2]
k = 4
max_sum = 0

for i in range(len(arr) - k + 1):
    current_sum = 0
    for j in range(i, i + k):
        current_sum += arr[j]
    max_sum = max(max_sum, current_sum)

print(max_sum,current_sum)   # 11
````

````python
arr = [3, 1, 4, 1, 5, 9, 2, 6]
k = 4
window_sum = sum(arr[:k])    
max_sum = window_sum
for i in range(k, len(arr)):
    window_sum += arr[i]       
    window_sum -= arr[i - k]   
    max_sum = max(max_sum, window_sum)

print(max_sum)   # 22
````
