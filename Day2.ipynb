# Day 2 — Time & Space Complexity (Big O Notation)

#### Time complexity ka matlab ye nahi ke code actually kitne seconds mein chalta hai — balke ye hai ke input size badhne ke saath code ke steps kitni tezi se badhte hain.
- Simple example:

Agar n = 5 hai to 5 baar chala. Agar n = 1000 hai to 1000 baar chala. Input ke saath linearly badh raha hai — isko O(n) kehte hain.

## Pehle mein extra memory nahi li — O(1) space. Doosre mein n items store kiye — O(n) space.
```
How to Analyze Your Own Code
Jab bhi code likho, khud se ye 3 sawaal poochho:

Kitne loops hain? — ek loop = O(n), nested loops = O(n²)
Kya input har baar half ho rahi hai? — O(log n)
Kya extra list/array bana rahe ho? — O(n) space
```

````python
n = 5
n = 100
for i in range(n):
    print(i)
````

#### The Most Common Big O Values (with feel)

| Big O     | Name         | Simple Meaning                                        |
|-----------|--------------|-------------------------------------------------------|
| O(1)      | Constant     | Input kitna bhi bado, steps same rehte hain           |
| O(log n)  | Logarithmic  | Har baar problem half ho jaati hai                    |
| O(n)      | Linear       | Input ke saath seedha badhta hai                      |
| O(n²)     | Quadratic    | Nested loop — n ke square ke barabar steps            |
| O(2ⁿ)     | Exponential  | Recursion mein — bohot tezi se badh jaata hai         |

## O(1) — Constant Time

````python
arr = [10, 20, 30, 40, 50]
print(arr[0])   # always 1 step, doesn't matter if list has 5 or 5 million items
````

### O(n) — Linear Time

````python
arr = [1, 2, 3, 4, 5]
for i in arr:
    print(i)     # runs n times
````

### O(n²) — Quadratic Time

````python
n = 3
for i in range(n):       # runs n times
    for j in range(n):   # for each i, runs n times again
        print(i, j)
````

## O(log n) — Logarithmic Time
Ye tab hota hai jab har step mein problem half ho jaaye. Classic example Binary Search hai (jo Day 11 mein seekhoge), lekin concept aaj samjho:

```
n = 16
Step 1: 16 → 8
Step 2: 8  → 4
Step 3: 4  → 2
Step 4: 2  → 1
```

# What is Space Complexity?
Space complexity ye batata hai ke code kitni extra memory use karta hai input size ke hisaab se.


````python
# O(1) space — no extra storage, just printing
for i in range(n):
    print(i)

# O(n) space — storing n items in a new list
arr = []
for i in range(n):
    arr.append(i)
````

## Pehle mein extra memory nahi li — O(1) space. Doosre mein n items store kiye — O(n) space.
```
How to Analyze Your Own Code

Kitne loops hain? — ek loop = O(n), nested loops = O(n²)
Kya input har baar half ho rahi hai? — O(log n)

```

````python
def get_first(arr): # its o(1) because it contain only one character no matter how much values are there
    return arr[0]
````

````python
def print_all(arr):
    for item in arr: # its o(n) number for iterations and number of values like how much values are there n also is n steps
        print(item)
````

````python
def find_sum(arr):
    total = 0
    for i in arr:
        total += i  # its is a space complexity o(n) because append the total value use memory 
    return total
````

````python
def print_pairs(arr):
    for i in arr:
        for j in arr:
            print(i, j) # o(n^2) because it use nested loop...
````
